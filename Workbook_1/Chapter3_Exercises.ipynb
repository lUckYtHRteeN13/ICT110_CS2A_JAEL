#### Exercise 1

In the code below, which will never print regardless of the value for x? Edit the code to correct the error.

```python
if x > 0 :
    print('Above 0')
elif x >  10 :
    print('Above 10')
elif x > 50 :
    print('Above 50')
elif x >  20 :
    print('Above 20')
else :
    print('Something else')
```

The code above will only print the value inside the first conditional, which is "Above 0". <br> Below is the correct code.

````python
x = input("Number: ")

try:
  x = int(x)

  if x < 0 :
    print('Above 0')
  elif x <  10 :
      print('Above 10')
  elif x < 50 :
      print('Above 50')
  elif x <  20 :
      print('Above 20')
  else :
      print('Something else')

except:
  print("Invalid input")
````

#### Exercise 2

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E3.2

````python
number_of_hours = input("Enter Hours: " )
rate = input("Enter Rate: " )

try:
  number_of_hours = int(number_of_hours)
  rate = int(rate)

  overtime_hours = max(0, number_of_hours - 40)

  number_of_hours = min(40, number_of_hours)
  overall_pay = rate * number_of_hours + (rate * 1.5 * overtime_hours)

  print(f"Pay: {overall_pay}")

except:
  print("invalid Input/s")
````

## CHALLENGE

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

    Score           Grade
    [ 97, 100]      1.0
    [ 93,  97)		1.25
    [ 89,  93)		1.5
    [ 85,  89)		1.75
    [ 81,  85)		2.0
    [ 77,  81)		2.25
    [ 74,  77)		2.5
    [ 0,   74)		3.0

Adapted from [PEW] E3.3
<br>

````python
grade = input("Score: ")

try:
  grade = int(grade)

  if grade > 100:
    print("Invalid")
  elif grade >= 97:
    print("Grade: 1.0")
  elif grade >= 93:
    print("Grade: 1.25")
  elif grade >= 89:
    print("Grade: 1.5")
  elif grade >= 85:
    print("Grade: 1.75")
  elif grade >= 81:
    print("Grade: 2.0")
  elif grade >= 77:
    print("Grade: 2.25")
  elif grade >= 74:
    print("Grade: 2.5")
  elif grade >= 0:
    print("Grade: 3.0")

except:
  print("Invalid Input!!")
````
