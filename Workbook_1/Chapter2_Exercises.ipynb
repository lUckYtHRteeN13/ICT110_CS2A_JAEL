#### Exercise 2

Write a program to prompt the user for hours and rate per hour to compute gross pay.
```
Enter Hours: 35
Enter Rate: 2.75
Pay: 96.25
```

[PEW] E2.3

````python
number_of_hours = int(input("Enter Hours: "))
rate_per_hour = float(input("Enter Rate: "))
pay = number_of_hours * rate_per_hour
print("Pay: " + str(pay))
````

#### Exercise 3

Assume that we execute the following assignment statements:
```python
width = 17
height = 12.0
```
For each of the following expressions, write the value of the expression and the type (of the value of the expression).
```python
width//2
width/2.0
height/3
1 + 2 * 5
```

[PEW] E2.4

````python
width = 17
height = 12.0

print(width//2 , type(width//2))
print(width/2.0, type(width/2.0))
print(height/3, type(height/3))
print(1 + 2 * 5, type(1 + 2 * 5))
````

ANSWERS
```python
width//2 = 8 (int)
width/2.0 = 8.5 (float)
height/3 = 4.0 (float)
1 + 2 * 5 = 11 (int)
```

## CHALLENGE

Write a program which prompts the user for a Celsius temperature, convert the temperature to Fahrenheit, and print out the converted temperature.

[PEW] E2.5

````python
celsius = float(input("Temperature (Celsius): "))
fahrenheit = celsius * (9/5) + 32
print(f"Fahrenheit: {fahrenheit}")
````
