#### Exercise 1

Investigate on the difference between `is` and `==` given the following variables.
```python
a = 2
b = 2.0
c = a
```

```python
a = 2 # is an int
b = 2.0 # is a float
c = a # reference to a
```

" == " checks the two variables equals to each other, otherwise false <br>
" is " checks the if the variable is the same object, otherwise false

````python
a = 2
b = 2.0
c = a

print(a == b)
print(a is b)
print(a == c)
print(a is c)

print(b == c)
print(b is c)
````

#### Challenge 1

Write a program which repeatedly reads numbers until the user enters "done". Once "done" is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number.

    Enter a number: 4
    Enter a number: 5
    Enter a number: bad data
    Invalid input
    Enter a number: 7
    Enter a number: done
    16 3 5.333333333333333
    
[PEW] E5.1

````python
count = 0
sum = 0

while True:
  user = input("Enter a number: ")

  if (user == "done"):
    print(sum, count, sum/count)
    break

  if (user.isdigit()):
    count+=1
    sum += int(user)

  else:
    print("Invalid input")
````

#### Challenge 2  

Write another program that prompts for a list of numbers as above and at the end prints out both the maximum and minimum of the numbers instead of the average.

[PEW] E5.2

````python
inputs = []
largest = -1
while True:
  try:

    user = input("Enter a number: ")
    if (user == "done"):
      for num in inputs:

        if num > largest:
          largest = num

      smallest = inputs[0]
      for i in range(1, len(inputs)):
        if (inputs[i] < smallest):
          smallest = inputs[i]

      print(largest, smallest)
      break

    inputs.append(int(user))

  except Exception as e:
    print("Invalid Input")
````
